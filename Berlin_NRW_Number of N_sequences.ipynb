```python
import pandas as pd
import csv

def read_fasta(fp):
        name, seq = None, []
        for line in fp:
            line = line.rstrip()
            if line.startswith(">"):
                if name: yield (name, ''.join(seq))
                name, seq = line, []
            else:
                seq.append(line)
        if name: yield (name, ''.join(seq))

seq_list =[]
n_count = []
with open("/content/sequences_Berlin_2021.fasta") as fp:
    for name, seq in read_fasta(fp):
      seq_list.append(name)
      n_count.append(seq.count('N'))

print(seq_list)
print(n_count)

import csv
  
  
# field names 
import pandas as pd  


# dictionary of lists  
dict = {'n_count': n_count, 'seq_count': seq_list}  
       
df = pd.DataFrame(dict) 
    
# saving the dataframe 
df.to_csv('Berlin.csv')
```

```python
def count_elements(seq) -> dict:
     """Tally elements from `seq`."""
     hist = {}
     for i in seq:
         hist[i] = hist.get(i, 0) + 1
     return hist
counted = count_elements(seq)
counted
```
